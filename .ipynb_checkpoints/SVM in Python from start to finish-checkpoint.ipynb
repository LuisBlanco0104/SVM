## About SVMs

**Support Vector Machines** are one of the best machine learning methods when getting the correct answer is a higher priority than understanding **why** you get the correct answer. They work really sell with relatively small datasets and they tendo to work well "out of the box". in other words, they do not require much optimization.


<center><img src="svm.PNG" width="400"/></center>

It is clear that we can easily divide the two classes by drawing a line in between them. The goal of support vector machines is to maximise the distance between those two classes to create maximal separation. What I mean is that we want to find the line that runs in between the two most extreme points of both sets and maximises the margins between both classes.

## 2. Identifying missing data 

Unfortunatly the biggest part of any data analaysis project is making sure that the data is correctly formatted and fixxing it when it is not. The first part of this process is identifying and dealing with the meassing data.

**Missing data** is simply a blank space or a surrogate value like **NA**, that indicates the we failed to collect data for one of the features. For example, if we forgot to ask someone's age or forgot to write it down, then we would have a blank space.

There are two ways of dealing with missing data:

   1. We can remove the rows that contain missing data. This is relatively easy to do but it wastes all of the other values that we collected. How big of a waste depends on how important the missing value is for the classification. 
   2. We can **impute** the values that are missing. In this context impute is just a fancy way of saying "we can make an educated guess about what the value should be".

## 4. Downsample the data

**Support Vector Machines** are great with small datasets, but not awesome with large ones, and this dataset, is relatively large, or big enough to take a long time to optimize with `Cross validation`. So we will downsample both categories, the customers who did and did not default to 1000.


Firs let's remind ourselves how many customers are there: **29,932**. 


So for that we will sepparate the data into thos who did not default (0) and those who did (1).

```python
df_no_default = dfn[dfn['DEFAULT']==0]
df_default = dfn[dfn['DEFAULT']==1]
```

    let's downsample

```python
df_no_default_1000 = resample(df_no_default,
                             replace=False,
                             n_samples=1000,
                             random_state=42)
len(df_no_default_1000)
```

```python
df_default_1000 = resample(df_default,
                           replace=False,
                           n_samples=1000,
                           random_state=42)
len(df_default_1000)
```

    Now letÂ´s merge the two datasets into one in order to have just one dataframe with 2000 values.

<div class="alert alert-block alert-success">
    <b>Note:</b> These last steps could be simplified in less code but for this being the first time, let's do it step by step
</div>


```python
df_downsapled = pd.concat([df_default_1000, df_no_default_1000])
```

## 5. Formatting the data for SVM: 
### Split data into X and y


Now that we have taken care of the missing data, we are ready to start formatting the data for **Support vector Machine**.

The first step is to split the data into:
1. `X` Independent Variables
2. `y` Dependent variable

```python
X = df_downsapled.drop(['DEFAULT'],axis=1).copy()
y = df_downsapled['DEFAULT'].copy()
```

### One-Hot Encoding

Here we have to take a look at the variables in `X` in order to have a **binary** categorical data (1, 0).

<div class="alert alert-block alert-warning">
    <b>Note:</b> Scickit-learn DT nativly supports continous data, like resting blood (restbp) but does NOT support categorical data like <b>EDUCATION </b> which contains 4 different categories. 
</div>


<div class="alert alert-block alert-success">
    <b>To do:</b> Thus, in order to use categorical data with <b>Scikit-learn  SVM</b> we have to convert the column of categorical data into multiple columns of binary values (0,1). This trick is called <b>One-Hot Encoding</b>   
</div>


At this point you may be wondering *WHat's wrong with treating categorical data like continous data?*. To answer that let's look at MARRIAGE:

    1. Married
    2. Single
    3. Other


If we treated these values 1,2,3 like continous data, then we would assume that 3 is more similar to 2 than it is to 1 (*Other* is more similar to *Single* than it is to *Married*). That means that SVM would be more likely to cluster the people with 3s and 2s together than people with 3s and 1s. In contrast, if we treat these numbers like categorical data, then we treat each one as a seperate category that is no more or less similar to any of the other categories. Thus, the likelihood of clustering people with 3s with 2s is the same as 3s with 1s. and that approach is more reasonoble.


We use the build in function of pandas `get_dummies`, what it does is to vonvert the categorical values of the one Column and transform it into several columns of binery values with the corresponding names, for example:


**EDUCATION** = 1 Column with values  1, 2, 3, 4  ---- > `get_dummies` ----> 4 columns with values 1, 0
named: **EDUCATION_1**, **EDUCATION_2**, **EDUCATION_3**, **EDUCATION_4**.

```python
cols = (['SEX', 'EDUCATION', 'MARRIAGE', 'PAY_0', 'PAY_2',
       'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6'])


X_encoded = pd.get_dummies(X, columns=cols)
X_encoded.head()
```

### Center and Scale the data

THe **Radial Basis Function (RBF)** that we will use w/ our SVM, assumes that the data are centered and scaled. In other words, each column should have a value $mean=0$ and a standard deviation $std = 1$. So we need to do this to both training and testing datasets.

<div class="alert alert-block alert-warning">
    <b>Note:</b> We split the data first and then center/scale them seperatly to avoid <a href=https://machinelearningmastery.com/data-leakage-machine-learning/ >Data Leakage</a>. Data Leakage occurs when information about the training dataset corrupts or influences the training dataset.
</div>

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size = 0.3, random_state=42)
X_train = scale(X_train)
X_test = scale(X_test)
```

***
<a id='first_svm'></a>
## 6. Building a preliminary SVM


At long last, the data is correctly formatted for making a **SVM**.X_train

```python
svm = SVC(random_state=42).fit(X_train, y_train)
```

    We have build the SVM now let's see how it performs.

```python
plot_confusion_matrix(svm,
                     X_test,
                     y_test,
                     values_format='d',
                     display_labels=['Did not default','Did default']);
print(svm.score(X_test, y_test))
```

The SVM produced by the default parameters is **meh!**, has an accuracy of **72%** but it can do better. So let's try to improve predictions using `cross_val` to optimize the parameters.

***
<a id='opt'></a>
## 7. Optimizing parameters with Cross-Validation


Optimizing a **Support Vector Machine** is all about finding the best value for `gamma` and, potentially the regularization parameter `C`. So let's see if we can find better parameter values using cross valudation in hope that we can improve the accuracy with the testing dataset.


Since we have to parameters to optimize, we will use `GridSearchCV()`. We specify a bunch of potential values for `gamma` and `C`, and `GridSearchCV()` tests all possible combinations of the parameters for us.

```python
# Default values are C=1 and gamma='scale', we include them because why not?
param_grid = [{'C': [0.5, 1, 10, 100], #C must be > 0
              'gamma':['scale', 1, 0.1, 0.001, 0.0001],
              'kernel':['rbf']}]
optimal_params = GridSearchCV(SVC(),
                             param_grid,
                             cv=5,
                             scoring='accuracy',
                             verbose=0)
optimal_params.fit(X_train, y_train)
print(optimal_params.best_params_)
```

<a id='final'></a>
And we see that the best value for C is 1, for gamma 0.001 and kenrel as we did not specify another one then 'rbf' for Radial Base Function. 

***

## Building, Evaluating, Drawing and Interpreting the Final Support Vector Machine

Now that we have the best params. We can build the **Final Support Vector Machine**




```python
svm_final = SVC(random_state=42, C=1, gamma=0.001, kernel='rbf').fit(X_train, y_train)
```

```python
plot_confusion_matrix(svm_final,
                     X_test,
                     y_test,
                     values_format='d',
                     display_labels=['Did not default','Did default']);
print(svm_final.score(X_test, y_test))
```

Let's now use <a href=https://de.wikipedia.org/wiki/Hauptkomponentenanalyse >**Principal Component Analysis**</a>. to plot the 24-dimensional dataset.


```python
pca = PCA()
X_pca= pca.fit_transform(X_train)
per_var = np.round(pca.explained_variance_ratio_*100, decimals=1)
labels = [str(x) for x in range(1, len(per_var) + 1)]

plt.bar(x=range(1, len(per_var)+1), height=per_var)
plt.tick_params(axis='x',
               which='both',
               bottom=False,
               labelbottom=False,
               top=False)
plt.ylabel('Percentage of Explained Variance')
plt.xlabel('Principal Component Analysis')
plt.title('Screen Plot')
plt.show()
```

```python
train_pc1 = X_pca[:, 0]
train_pc2 = X_pca[:, 1]

pca_train_scaled= scale(np.column_stack((train_pc1, train_pc2)))

param_grid = [{'C': [0.5, 1, 10, 100, 1000], #C must be > 0
              'gamma':['scale', 1, 0.1, 0.001, 0.0001],
              'kernel':['rbf']}]

optimal_params = GridSearchCV(SVC(),
                             param_grid,
                             cv=5,
                             scoring='accuracy',
                             verbose=0)
optimal_params.fit(X_train, y_train)
print(optimal_params.best_params_)
```

```python
svm = SVC(random_state=42, C=1, gamma=0.001).fit(pca_train_scaled, y_train)

X_test_pca = pca.transform(X_train)
test_pc1 = X_test_pca[:, 0]
test_pc2 = X_test_pca[:, 1]

x_min = test_pc1.min() - 1
x_max = test_pc1.max() + 1
y_min = test_pc2.min() - 1
y_max = test_pc2.max() + 1

xx, yy = np.meshgrid(np.arange(start=x_min, stop=x_max, step=0.1),
                     np.arange(start=y_min, stop=y_max, step=0.1))

Z = svm.predict(np.column_stack((xx.ravel(), yy.ravel())))

Z = Z.reshape(xx.shape)


fig, ax = plt.subplots(figsize=(10,10))
ax.contourf(xx, yy, Z, alpha=0.1)
cmap = colors.ListedColormap(['#e41a1c', '#4daf4a'])
scatter = ax.scatter(test_pc1, test_pc2, c=y_train,
                    cmap= cmap,
                    s=100,
                    edgecolors='k',
                    alpha=0.7)
legend = ax.legend(scatter.legend_elements()[0],
                  scatter.legend_elements()[1],
                  loc='upper right')
legend.get_texts()[0].set_text('No Default')
legend.get_texts()[1].set_text('Yes Default')
ax.set_ylabel('PC2')
ax.set_xlabel('PC1')
ax.set_title('Decision surface using PCA transformed/projected features')
plt.show()
```
